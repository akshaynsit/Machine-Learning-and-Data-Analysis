# Haberman's Survival Dataset
The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.



Dataset taken from kaggle : https://www.kaggle.com/gilsousa/habermans-survival-data-set


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Load Haberman survival data stored from CSV into a pandas dataFrame.
haberman = pd.read_csv("haberman.csv")
print(haberman)
```

```python
#naming the column name for the dataset
haberman.columns= ['age', 'year_of_operation', 'lymph_nodes', 'surv_status']
print(haberman)
```

```python
# Q - how many data-points and features available in haberman dateset ?
print ("Shape of Haberman : ",haberman.shape)
```

```python
# Q - write down the column names present in haberman dataset
print("Name of the columns :- ",haberman.columns)
```

```python
#converting integer values of surv_status into string type 1 --> "Yes" and 2 --> "No"
haberman.surv_status = list(map(lambda x : "Yes" if x == 1 else "No" , haberman.surv_status))
```

```python
#(Q) How many data points for each class are present? 

print(haberman.surv_status.value_counts()) 
# this show haberman datasets are imblanced in nautre becasue survival ratio is in 224(Yes):81(N0)
```

**Obervation** **Till now**                                                                                                                  

1) This dataset have 305 rows and 4 columns.                                                                           
2) Names of columns are :- 'age', 'year_of_operation', 'lymph_nodes', 'surv_status'.                                             
3) Dataset have 3 independent variable and 1 dependent variable(**surv_status**) .                                                   
4) Dependent variable have two types of values(Yes/No) :  yes = Patients will survive for 5 year or longer , No = Patients died before 5 year .                                                                                                                 
5) Haberman is imbalance datasets as the raito of class1(yes) and class2(No) is 224:81

```python
#1-D scatter plot of petal-length
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

surv_status_yes = haberman.loc[haberman["surv_status"] == "Yes"];
surv_status_no = haberman.loc[haberman["surv_status"] == "No"];

plt.plot(surv_status_yes["lymph_nodes"],np.zeros_like(surv_status_yes["lymph_nodes"]),'s')
plt.plot(surv_status_no["lymph_nodes"],np.zeros_like(surv_status_no["lymph_nodes"]),'o')

plt.show()

#'age', 'year_of_operation', 'lymph_nodes', 'surv_status'
```
